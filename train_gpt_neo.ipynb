```python
from happytransformer import GENSettings

# https://happytransformer.com/text-generation/settings/
args= GENSettings(no_repeat_ngram_size=2, do_sample=True, top_k=50, temperature=0.7, max_length=200)

result = happy_gen.generate_text("Je pense donc je suis", args=args)
print(result.text)
```

```python
import json
with open('../French-Politic-Speech-Dataset/speechs_data.json', 'r') as f:
    data = json.load(f)['speechs']
    data = [d for d in data if d['kind'] == 'Texte intégral']

print(len(data))
```

```python
N = 10 # default -1
J = 1000 # default 0

txt = ''
subset = data[J:J+N] if N > 0 else data
for s in subset:
    if s['text'] and s['title']:
        txt += s['title'] + '\n' + s['text'] + '\n\n'

filename = f"./texts_{N}.txt" if N > 0 else "./texts_full.txt"
with open(filename, 'w+') as f:
    f.write(txt)
```

```python
happy_gen.train("./texts_1000.txt")
happy_gen.save("./model_125M_1000S.txt")
```
